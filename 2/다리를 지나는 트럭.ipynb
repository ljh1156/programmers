```python
def solution(bridge_length, weight, truck_weights):
    answer = 0

    
    on_bridge = []
    bridge_count_dic = {}
    while(1):
        print("{}:{}:{}".format(answer, on_bridge, truck_weights))
        answer += 1

        if on_bridge:
            for i in range(0, len(on_bridge)):
                on_bridge[i][1] -= 1
            if on_bridge[0][1] == 0:
                on_bridge.pop(0)
        
        if len(truck_weights) == 0 and len(on_bridge) == 0:
            break
                    
        if truck_weights:
            on_bridge_weight = 0
            for on_b in on_bridge:
                on_bridge_weight += on_b[0]
            if on_bridge_weight + truck_weights[0] <= weight \
                and len(on_bridge) + 1 <= bridge_length:
                on_bridge.append([truck_weights[0], bridge_length])
                truck_weights.pop(0)
 
    return answer
```
